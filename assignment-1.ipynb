```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the dataset


```python
df = pd.read_csv("data/swedish_population_by_year_and_sex_1860-2022.csv")
df
```

```python
df.describe()
```

## Part I: Dependency ratio

Creating a new data frame containing the amount of people of a certain age for each year:

```python
age_df = df.drop(columns='sex').groupby("age").sum()
age_df
```

Generating a support vector with all the ages in the dataset:

```python
age = pd.to_numeric(age_df.index.str.replace("+", ""))
```

Calculating the amount of children by summing every age under 15 years old:

```python
children_count = age_df[age <= 14].sum()
children_count
```

Doing the same thing for the elderly:

```python
elderly_count = age_df[age >= 65].sum()
elderly_count
```

And for the labor force:

```python
labor_count = age_df[(age >= 15) & (age <= 64)].sum()
labor_count
```

Calculating the dependency ratio as: $\frac{\text{children\_count} + \text{elderly\_count}}{\text{labor\_count}}$ 

```python
dependent_sum = children_count + elderly_count
dependency_ratio = dependent_sum / labor_count
dependency_ratio
```

Plotting the ratio:

```python
fig, ax = plt.subplots()
ax.plot(dependency_ratio, label="Dependency ratio")
ax.set_xlabel('Year')
ax.set_ylabel('Dependency Ratio')
ax.set_xticks(range(0, len(dependency_ratio), 20))
ax.legend()

plt.show()
```

### Part II: Population fraction


```python
children_fraction = children_count / dependent_sum
elderly_fraction = elderly_count / dependent_sum
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
ax[0].plot(children_fraction)
ax[0].set_title('Children Fraction')
ax[0].set_xlabel('Year')
ax[0].set_ylabel('Fraction')
ax[0].set_xticks(range(0, len(children_fraction), 20))
ax[1].plot(elderly_fraction)
ax[1].set_title('Elderly Fraction')
ax[1].set_xlabel('Year')
ax[1].set_ylabel('Fraction')
ax[1].set_xticks(range(0, len(elderly_fraction), 20))
plt.show()
```

### Part (ii) Alternative

```python
total_sum = age_df.sum()
children_fraction = children_count / total_sum
elderly_fraction = elderly_count / total_sum
dependent_fraction = dependent_sum / total_sum
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].plot(children_fraction)
ax[0].set_title('Children Fraction')
ax[0].set_xlabel('Year')
ax[0].set_ylabel('Fraction')
ax[0].set_xticks(range(0, len(children_fraction), 20))
ax[1].plot(elderly_fraction)
ax[1].set_title('Elderly Fraction')
ax[1].set_xlabel('Year')
ax[1].set_ylabel('Fraction')
ax[1].set_xticks(range(0, len(elderly_fraction), 20))
ax[2].plot(dependent_fraction)
ax[2].set_title('Dependent Fraction')
ax[2].set_xlabel('Year')
ax[2].set_ylabel('Fraction')
ax[2].set_xticks(range(0, len(dependent_fraction), 20))
plt.show()
```
